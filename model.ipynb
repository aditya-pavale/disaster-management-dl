```python
# ================================
# MULTIMODAL DENSENET TRAINING WITH DUPLICATE DETECTION
# Disaster Classification - DenseNet Version
# ================================

import os
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from sklearn.neighbors import NearestNeighbors
import pickle
from collections import defaultdict, deque

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models
from torch.amp import autocast, GradScaler

# =====================================================
# GLOBAL DEVICE (GPU / CPU)
# =====================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
use_amp = device.type == "cuda"  # use mixed precision only on GPU
scaler = GradScaler('cuda', enabled=use_amp)
print(f"Using device: {device}")

# =====================================================
# VS CODE FILE PATHS
# =====================================================
TRAIN_CSV = "train_clip_results.csv"
VAL_CSV = "val_clip_results.csv"
TEST_CSV = "test_clip_results.csv"
IMAGES_ROOT = "images"

WORK = "."
TRAIN_DIR = "train"
VAL_DIR = "val"
TEST_DIR = "test"

# =====================================================
# LOAD CSVs
# =====================================================
train_df = pd.read_csv(TRAIN_CSV)
val_df = pd.read_csv(VAL_CSV)
test_df = pd.read_csv(TEST_CSV)

# Normalize filename column
for df in (train_df, val_df, test_df):
    if "image" in df.columns:
        df.rename(columns={"image": "file_name"}, inplace=True)

# =====================================================
# CLASS LABELS
# =====================================================
class_names = sorted(train_df["label"].unique())
class2idx = {c: i for i, c in enumerate(class_names)}
idx2class = {i: c for c, i in class2idx.items()}

train_df["class_idx"] = train_df["label"].map(class2idx)
val_df["class_idx"] = val_df["label"].map(class2idx)
test_df["class_idx"] = test_df["label"].map(class2idx)

print("Classes:", class2idx)
print("Class distribution - Train:", train_df["label"].value_counts().to_dict())
print("Class distribution - Val:", val_df["label"].value_counts().to_dict())
print("Class distribution - Test:", test_df["label"].value_counts().to_dict())

# =====================================================
# COPY IMAGES TO WORKING DIRECTORIES
# =====================================================
print("Copying images to working directories...")
os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(VAL_DIR, exist_ok=True)
os.makedirs(TEST_DIR, exist_ok=True)

def copy_images(df, target_dir):
    """Copy images to target directory with proper error handling"""
    copied = 0
    missing = 0
    for fname in tqdm(df["file_name"], desc=f"Copying to {target_dir}"):
        src = os.path.join(IMAGES_ROOT, fname)
        dst = os.path.join(target_dir, fname)
        
        # Create subdirectories if needed
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        
        if os.path.exists(src):
            if not os.path.exists(dst):  # Don't copy if already exists
                shutil.copy2(src, dst)
            copied += 1
        else:
            missing += 1
            print(f"Missing source: {src}")
    
    print(f"Copied {copied} images, {missing} missing")
    return copied, missing

# Copy images for each split
copy_images(train_df, TRAIN_DIR)
copy_images(val_df, VAL_DIR)
copy_images(test_df, TEST_DIR)

# =====================================================
# DUPLICATE DETECTION SYSTEM
# =====================================================
class DuplicateDetector:
    def __init__(self, threshold=0.85, device=device):
        self.threshold = threshold
        self.embedding_model = None
        self.nn_model = None
        self.embeddings = None
        self.image_paths = None
        self.device = device

    def build_embeddings(self, image_dir, image_files):
        """Build image embeddings using DenseNet121"""
        print("Building image embeddings for duplicate detection...")

        # Use pre-trained DenseNet121 for embeddings
        if self.embedding_model is None:
            self.embedding_model = models.densenet121(
                weights=models.DenseNet121_Weights.IMAGENET1K_V1
            )
            self.embedding_model.classifier = nn.Identity()
            self.embedding_model.eval()
            self.embedding_model.to(self.device)

        # Image transforms for embedding model
        transform = transforms.Compose([
            transforms.Resize(256),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406],
                                 [0.229, 0.224, 0.225])
        ])

        embeddings = []
        valid_paths = []

        for img_file in tqdm(image_files):
            img_path = os.path.join(image_dir, img_file)
            try:
                img = Image.open(img_path).convert('RGB')
                img_tensor = transform(img).unsqueeze(0).to(self.device, non_blocking=True)

                with torch.no_grad(), autocast('cuda', enabled=(self.device.type == "cuda")):
                    embedding = self.embedding_model(img_tensor)
                    embedding = embedding.detach().cpu().numpy().flatten()

                embeddings.append(embedding)
                valid_paths.append(img_file)
            except Exception as e:
                print(f"Error processing {img_file}: {e}")
                continue

        self.embeddings = np.array(embeddings)
        self.image_paths = valid_paths

        # Build nearest neighbors model
        self.nn_model = NearestNeighbors(n_neighbors=10, metric='cosine')
        self.nn_model.fit(self.embeddings)
        print(f"Built embeddings for {len(self.embeddings)} images")

    def find_duplicate_groups(self):
        """Find duplicate groups using union-find algorithm"""
        if self.nn_model is None:
            raise ValueError("Build embeddings first!")

        print("Finding duplicate groups...")

        # Union-Find data structure
        parent = {i: i for i in range(len(self.image_paths))}

        def find(x):
            while parent[x] != x:
                parent[x] = parent[parent[x]]
                x = parent[x]
            return x

        def union(x, y):
            parent[find(x)] = find(y)

        # Find neighbors for each image
        distances, indices = self.nn_model.kneighbors(self.embeddings)

        # Connect similar images
        for i in range(len(self.image_paths)):
            for j, dist in zip(indices[i], distances[i]):
                if i != j and dist < (1 - self.threshold):
                    union(i, j)

        # Group images by connected components
        groups = defaultdict(list)
        for i in range(len(self.image_paths)):
            root = find(i)
            groups[root].append(self.image_paths[i])

        # Filter groups with duplicates
        duplicate_groups = {k: v for k, v in groups.items() if len(v) > 1}

        print(f"Found {len(duplicate_groups)} duplicate groups")
        for i, (key, group) in enumerate(list(duplicate_groups.items())[:5]):
            print(f"Group {i+1}: {len(group)} images - {group[:3]}...")

        return duplicate_groups

    def detect_duplicates_in_dataset(self, df_list, df_names):
        """Detect duplicates across all datasets"""
        all_files = []
        all_dfs = []
        df_indices = []

        # Collect all files
        for df, name in zip(df_list, df_names):
            files = df['file_name'].tolist()
            all_files.extend(files)
            all_dfs.extend([name] * len(files))
            df_indices.extend(range(len(files)))

        # Build embeddings for all files
        self.build_embeddings(IMAGES_ROOT, all_files)

        # Find duplicate groups
        duplicate_groups = self.find_duplicate_groups()

        # Map groups to original dataframes
        file_to_df = dict(zip(all_files, all_dfs))
        file_to_idx = dict(zip(all_files, df_indices))

        return duplicate_groups, file_to_df, file_to_idx

# =====================================================
# APPLY DUPLICATE DETECTION
# =====================================================
print("=" * 60)
print("DUPLICATE DETECTION AND DATA SPLITTING")
print("=" * 60)

detector = DuplicateDetector(threshold=0.85, device=device)
duplicate_groups, file_to_df, file_to_idx = detector.detect_duplicates_in_dataset(
    [train_df, val_df, test_df],
    ['train', 'val', 'test']
)

# Force all duplicates into training set
print("\nMoving duplicates to training set...")
moved_count = 0

# Create copies to avoid modifying during iteration
train_df_updated = train_df.copy()
val_df_updated = val_df.copy()
test_df_updated = test_df.copy()

for group_id, files in duplicate_groups.items():
    group_dfs = [file_to_df[f] for f in files]

    # If group spans multiple splits, move all to train
    if len(set(group_dfs)) > 1:
        for file in files:
            original_split = file_to_df[file]
            if original_split != 'train':
                if original_split == 'val':
                    file_data = val_df_updated[val_df_updated['file_name'] == file]
                    if len(file_data) > 0:
                        val_df_updated = val_df_updated[val_df_updated['file_name'] != file]
                        if file not in train_df_updated['file_name'].values:
                            train_df_updated = pd.concat([train_df_updated, file_data],
                                                         ignore_index=True)
                            moved_count += 1

                elif original_split == 'test':
                    file_data = test_df_updated[test_df_updated['file_name'] == file]
                    if len(file_data) > 0:
                        test_df_updated = test_df_updated[test_df_updated['file_name'] != file]
                        if file not in train_df_updated['file_name'].values:
                            train_df_updated = pd.concat([train_df_updated, file_data],
                                                         ignore_index=True)
                            moved_count += 1

print(f"Moved {moved_count} images to training set to avoid data leakage")

# Update the dataframes
train_df = train_df_updated
val_df = val_df_updated
test_df = test_df_updated

print(f"Final dataset sizes:")
print(f"Train: {len(train_df)} images")
print(f"Val: {len(val_df)} images")
print(f"Test: {len(test_df)} images")

# =====================================================
# ENHANCED DATASET CLASS
# =====================================================
class DisasterDataset(Dataset):
    def __init__(self, df, directory, transform=None):
        self.df = df.reset_index(drop=True)
        self.dir = directory
        self.transform = transform
        
        # Filter out missing files and create mapping
        self.valid_indices = []
        self.valid_files = []
        self.valid_labels = []
        
        for idx in range(len(self.df)):
            fname = self.df.loc[idx, "file_name"]
            img_path = os.path.join(self.dir, fname)
            if os.path.exists(img_path):
                self.valid_indices.append(idx)
                self.valid_files.append(fname)
                self.valid_labels.append(self.df.loc[idx, "class_idx"])
            else:
                print(f"Warning: Image not found - {img_path}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        fname = self.valid_files[idx]
        label = self.valid_labels[idx]

        img_path = os.path.join(self.dir, fname)
        img = Image.open(img_path).convert("RGB")
        if self.transform:
            img = self.transform(img)

        return img, label

# =====================================================
# STRONGER TRANSFORMS WITH HEAVY AUGMENTATION
# =====================================================
train_tf = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.5, 1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.3),
    transforms.RandomApply([
        transforms.ColorJitter(0.4, 0.4, 0.4, 0.2)
    ], p=0.8),
    transforms.RandomRotation(20),
    transforms.RandomGrayscale(p=0.2),
    transforms.RandomApply([
        transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 2.0))
    ], p=0.3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.4, scale=(0.02, 0.3), value='random')
])

test_tf = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

train_ds = DisasterDataset(train_df, TRAIN_DIR, train_tf)
val_ds = DisasterDataset(val_df, VAL_DIR, test_tf)
test_ds = DisasterDataset(test_df, TEST_DIR, test_tf)

print(f"Valid training samples: {len(train_ds)}")
print(f"Valid validation samples: {len(val_ds)}")
print(f"Valid test samples: {len(test_ds)}")

# =====================================================
# CLASS WEIGHTED SAMPLER FOR IMBALANCED DATA
# =====================================================
train_labels = train_ds.valid_labels
class_weights = compute_class_weight('balanced',
                                     classes=np.unique(train_labels),
                                     y=train_labels)
sample_weights = [class_weights[label] for label in train_labels]

sampler = WeightedRandomSampler(sample_weights,
                                len(sample_weights),
                                replacement=True)

# On Windows, num_workers=0 avoids multiprocessing issues
loader_kwargs = {
    "batch_size": 32,
    "num_workers": 0,
    "pin_memory": (device.type == "cuda")
}

train_loader = DataLoader(train_ds, sampler=sampler, **loader_kwargs)
val_loader = DataLoader(val_ds, shuffle=False, **loader_kwargs)
test_loader = DataLoader(test_ds, shuffle=False, **loader_kwargs)

# =====================================================
# MULTIMODAL DENSENET MODEL WITH REGULARIZATION
# =====================================================
class MultimodalDenseNet(nn.Module):
    def __init__(self, num_classes, dropout_rate=0.6):
        super().__init__()

        # DenseNet backbone
        self.backbone = models.densenet121(
            weights=models.DenseNet121_Weights.IMAGENET1K_V1
        )

        # Freeze early layers
        for param in list(self.backbone.parameters())[:150]:
            param.requires_grad = False

        # Get feature dimension
        in_features = self.backbone.classifier.in_features

        # Enhanced classifier head with more regularization
        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(in_features, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate),
            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate * 0.7),
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate * 0.5),
            nn.Linear(256, num_classes)
        )

        # Replace the original classifier
        self.backbone.classifier = nn.Identity()

    def forward(self, x):
        features = self.backbone(x)
        output = self.classifier(features)
        return output

    def unfreeze_more_layers(self, epoch):
        # Gradual unfreezing strategy
        if epoch > 5:
            for param in list(self.backbone.parameters())[-100:]:
                param.requires_grad = True
        if epoch > 10:
            for param in list(self.backbone.parameters())[-200:]:
                param.requires_grad = True
        if epoch > 15:
            for param in self.backbone.parameters():
                param.requires_grad = True

model = MultimodalDenseNet(len(class_names), dropout_rate=0.6)
model = model.to(device)

# Label smoothing loss
class_weights_tensor = torch.tensor(class_weights,
                                    dtype=torch.float32,
                                    device=device)

class LabelSmoothCrossEntropyLoss(nn.Module):
    def __init__(self, weight=None, smoothing=0.1):
        super().__init__()
        self.weight = weight
        self.smoothing = smoothing

    def forward(self, pred, target):
        log_probs = torch.nn.functional.log_softmax(pred, dim=-1)
        nll_loss = -log_probs.gather(dim=-1, index=target.unsqueeze(1))
        nll_loss = nll_loss.squeeze(1)
        smooth_loss = -log_probs.mean(dim=-1)
        loss = (1 - self.smoothing) * nll_loss + self.smoothing * smooth_loss
        if self.weight is not None:
            loss = loss * self.weight[target]
        return loss.mean()

criterion = LabelSmoothCrossEntropyLoss(weight=class_weights_tensor,
                                        smoothing=0.1)

# Optimizer with different learning rates for backbone and classifier
backbone_params = []
classifier_params = []

for name, param in model.named_parameters():
    if 'backbone' in name and param.requires_grad:
        backbone_params.append(param)
    else:
        classifier_params.append(param)

optimizer = optim.AdamW([
    {'params': backbone_params, 'lr': 1e-5},  # Lower LR for backbone
    {'params': classifier_params, 'lr': 5e-4}  # Higher LR for classifier
], weight_decay=5e-4)

scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
    optimizer,
    T_0=10,
    T_mult=1,
    eta_min=1e-6
)
max_grad_norm = 1.0

# =====================================================
# TRAINING LOOP
# =====================================================
train_losses = []
val_losses = []
train_accs = []
val_accs = []

BEST_VAL_ACC = 0.0
BEST_VAL_LOSS = float('inf')
BEST_PTH_PATH = "best_multimodal_densenet.pth"
BEST_PKL_PATH = "best_multimodal_densenet.pkl"
PATIENCE = 12  # More patience for DenseNet
epochs_no_improve = 0

def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total, correct, loss_sum = 0, 0, 0

    pbar = tqdm(loader, desc="Training", leave=False)
    for X, y in pbar:
        X = X.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        optimizer.zero_grad()

        with autocast('cuda', enabled=use_amp):
            out = model(X)
            loss = criterion(out, y)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)
        scaler.step(optimizer)
        scaler.update()

        loss_sum += loss.item() * X.size(0)
        preds = torch.argmax(out, 1)
        correct += (preds == y).sum().item()
        total += y.size(0)

        pbar.set_postfix({'loss': loss.item(), 'acc': correct / total})

    return loss_sum / total, correct / total

def evaluate_epoch(model, loader, criterion, device):
    model.eval()
    total, correct, loss_sum = 0, 0, 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for X, y in loader:
            X = X.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)

            with autocast('cuda', enabled=use_amp):
                out = model(X)
                loss = criterion(out, y)

            loss_sum += loss.item() * X.size(0)
            preds = torch.argmax(out, 1)
            correct += (preds == y).sum().item()
            total += y.size(0)

            all_preds.extend(preds.detach().cpu().numpy())
            all_labels.extend(y.detach().cpu().numpy())

    return loss_sum / total, correct / total, all_preds, all_labels

print("Starting multimodal DenseNet training with duplicate-aware splits...")
for epoch in range(50):  # More epochs for DenseNet
    model.unfreeze_more_layers(epoch)
    
    train_loss, train_acc = train_epoch(model, train_loader, optimizer, criterion, device)
    val_loss, val_acc, _, _ = evaluate_epoch(model, val_loader, criterion, device)

    scheduler.step(epoch)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)

    current_lr_backbone = optimizer.param_groups[0]['lr']
    current_lr_classifier = optimizer.param_groups[1]['lr']
    
    print(
        f"Epoch {epoch + 1:2d} | "
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | "
        f"LR Backbone: {current_lr_backbone:.2e} | LR Classifier: {current_lr_classifier:.2e}"
    )

    improvement = val_acc > BEST_VAL_ACC + 0.001 or val_loss < BEST_VAL_LOSS - 0.002
    if improvement:
        BEST_VAL_ACC = max(BEST_VAL_ACC, val_acc)
        BEST_VAL_LOSS = min(BEST_VAL_LOSS, val_loss)
        epochs_no_improve = 0

        torch.save(model.state_dict(), BEST_PTH_PATH)
        with open(BEST_PKL_PATH, 'wb') as f:
            pickle.dump(model, f)
        print("  ↳ ✅ Saved new best model!")
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= PATIENCE:
            print(f"🛑 Early stopping after {epoch + 1} epochs!")
            break

    if epoch > 15 and (train_acc - val_acc) > 0.15:
        print(f"🛑 Stopping due to severe overfitting (gap: {train_acc - val_acc:.3f})")
        break

# =====================================================
# SAVE RESULTS
# =====================================================
training_results = {
    'model_state_dict': model.state_dict(),
    'class_names': class_names,
    'class2idx': class2idx,
    'idx2class': idx2class,
    'train_losses': train_losses,
    'val_losses': val_losses,
    'train_accs': train_accs,
    'val_accs': val_accs,
    'best_val_acc': BEST_VAL_ACC,
    'best_val_loss': BEST_VAL_LOSS,
    'duplicate_groups': duplicate_groups
}

with open("complete_training_results_densenet.pkl", 'wb') as f:
    pickle.dump(training_results, f)

# =====================================================
# PLOTS
# =====================================================
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label="Train Loss", linewidth=2, alpha=0.8)
plt.plot(val_losses, label="Val Loss", linewidth=2, alpha=0.8)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('DenseNet - Loss Curve')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(train_accs, label="Train Acc", linewidth=2, alpha=0.8)
plt.plot(val_accs, label="Val Acc", linewidth=2, alpha=0.8)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('DenseNet - Accuracy Curve')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("training_curves_densenet.png", dpi=150, bbox_inches='tight')
plt.show()

# =====================================================
# TEST EVALUATION
# =====================================================
print("\n" + "=" * 60)
print("FINAL TEST EVALUATION - DENSENET")
print("=" * 60)

model.load_state_dict(torch.load(BEST_PTH_PATH, map_location=device))
model.to(device)
model.eval()

test_loss, test_acc, y_true, y_pred = evaluate_epoch(model, test_loader, criterion, device)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Save test results
test_results = {
    'y_true': y_true,
    'y_pred': y_pred,
    'test_accuracy': test_acc,
    'test_loss': test_loss
}

with open("test_results_densenet.pkl", 'wb') as f:
    pickle.dump(test_results, f)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('DenseNet - Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("confusion_matrix_densenet.png", dpi=150, bbox_inches='tight')
plt.show()

# =====================================================
# FINAL ANALYSIS
# =====================================================
final_gap = train_accs[-1] - val_accs[-1] if train_accs and val_accs else 0.0
print(f"\n" + "=" * 60)
print("DENSENET TRAINING SUMMARY")
print("=" * 60)
print(f"✅ Duplicate groups found: {len(duplicate_groups)}")
print(f"✅ Images moved to train: {moved_count}")
print(f"✅ Best validation accuracy: {BEST_VAL_ACC:.4f}")
print(f"✅ Test accuracy: {test_acc:.4f}")
print(f"✅ Final overfitting gap: {final_gap:.4f}")

print(f"\n📁 FILES SAVED:")
print(f"  - {BEST_PTH_PATH} (PyTorch model)")
print(f"  - {BEST_PKL_PATH} (Pickle model)")
print(f"  - complete_training_results_densenet.pkl (Training data)")
print(f"  - test_results_densenet.pkl (Test results)")
print(f"  - training_curves_densenet.png (Plots)")
print(f"  - confusion_matrix_densenet.png (Confusion matrix)")

print(f"\n🎯 DENSENET SPECIFIC FEATURES:")
print(f"  - Enhanced classifier head (1024→512→256)")
print(f"  - Differential learning rates")
print(f"  - Gradual unfreezing strategy")
print(f"  - Higher capacity for complex features")
```

```python
# ================================
# MULTIMODAL TRAINING WITH DUPLICATE DETECTION
# Disaster Classification - No Overfitting + Duplicate Handling
# ================================

import os
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from sklearn.neighbors import NearestNeighbors
import pickle
from collections import defaultdict, deque

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models
from torch.amp import autocast, GradScaler

# =====================================================
# GLOBAL DEVICE (GPU / CPU)
# =====================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
use_amp = device.type == "cuda"  # use mixed precision only on GPU
scaler = GradScaler('cuda', enabled=use_amp)
print(f"Using device: {device}")

# =====================================================
# VS CODE FILE PATHS
# =====================================================
TRAIN_CSV = "train_clip_results.csv"
VAL_CSV = "val_clip_results.csv"
TEST_CSV = "test_clip_results.csv"
IMAGES_ROOT = "images"

WORK = "."
TRAIN_DIR = "train"
VAL_DIR = "val"
TEST_DIR = "test"

# =====================================================
# LOAD CSVs
# =====================================================
train_df = pd.read_csv(TRAIN_CSV)
val_df = pd.read_csv(VAL_CSV)
test_df = pd.read_csv(TEST_CSV)

# Normalize filename column
for df in (train_df, val_df, test_df):
    if "image" in df.columns:
        df.rename(columns={"image": "file_name"}, inplace=True)

# =====================================================
# CLASS LABELS
# =====================================================
class_names = sorted(train_df["label"].unique())
class2idx = {c: i for i, c in enumerate(class_names)}
idx2class = {i: c for c, i in class2idx.items()}

train_df["class_idx"] = train_df["label"].map(class2idx)
val_df["class_idx"] = val_df["label"].map(class2idx)
test_df["class_idx"] = test_df["label"].map(class2idx)

print("Classes:", class2idx)
print("Class distribution - Train:", train_df["label"].value_counts().to_dict())
print("Class distribution - Val:", val_df["label"].value_counts().to_dict())
print("Class distribution - Test:", test_df["label"].value_counts().to_dict())

# =====================================================
# COPY IMAGES TO WORKING DIRECTORIES (FIXED)
# =====================================================
print("Copying images to working directories...")
os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(VAL_DIR, exist_ok=True)
os.makedirs(TEST_DIR, exist_ok=True)

def copy_images(df, target_dir):
    """Copy images to target directory with proper error handling"""
    copied = 0
    missing = 0
    for fname in tqdm(df["file_name"], desc=f"Copying to {target_dir}"):
        src = os.path.join(IMAGES_ROOT, fname)
        dst = os.path.join(target_dir, fname)
        
        # Create subdirectories if needed
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        
        if os.path.exists(src):
            if not os.path.exists(dst):  # Don't copy if already exists
                shutil.copy2(src, dst)
            copied += 1
        else:
            missing += 1
            print(f"Missing source: {src}")
    
    print(f"Copied {copied} images, {missing} missing")
    return copied, missing

# Copy images for each split
copy_images(train_df, TRAIN_DIR)
copy_images(val_df, VAL_DIR)
copy_images(test_df, TEST_DIR)

# =====================================================
# DUPLICATE DETECTION SYSTEM
# =====================================================
class DuplicateDetector:
    def __init__(self, threshold=0.85, device=device):
        self.threshold = threshold
        self.embedding_model = None
        self.nn_model = None
        self.embeddings = None
        self.image_paths = None
        self.device = device

    def build_embeddings(self, image_dir, image_files):
        """Build image embeddings using DenseNet121 (runs on GPU if available)"""
        print("Building image embeddings for duplicate detection...")

        # Use pre-trained DenseNet121 for embeddings
        if self.embedding_model is None:
            self.embedding_model = models.densenet121(
                weights=models.DenseNet121_Weights.IMAGENET1K_V1
            )
            self.embedding_model.classifier = nn.Identity()  # Remove classifier for embeddings
            self.embedding_model.eval()
            self.embedding_model.to(self.device)

        # Image transforms for embedding model
        transform = transforms.Compose([
            transforms.Resize(256),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406],
                                 [0.229, 0.224, 0.225])
        ])

        embeddings = []
        valid_paths = []

        for img_file in tqdm(image_files):
            img_path = os.path.join(image_dir, img_file)
            try:
                img = Image.open(img_path).convert('RGB')
                img_tensor = transform(img).unsqueeze(0).to(self.device, non_blocking=True)

                with torch.no_grad(), autocast('cuda', enabled=(self.device.type == "cuda")):
                    embedding = self.embedding_model(img_tensor)
                    embedding = embedding.detach().cpu().numpy().flatten()

                embeddings.append(embedding)
                valid_paths.append(img_file)
            except Exception as e:
                print(f"Error processing {img_file}: {e}")
                continue

        self.embeddings = np.array(embeddings)
        self.image_paths = valid_paths

        # Build nearest neighbors model
        self.nn_model = NearestNeighbors(n_neighbors=10, metric='cosine')
        self.nn_model.fit(self.embeddings)
        print(f"Built embeddings for {len(self.embeddings)} images")

    def find_duplicate_groups(self):
        """Find duplicate groups using union-find algorithm"""
        if self.nn_model is None:
            raise ValueError("Build embeddings first!")

        print("Finding duplicate groups...")

        # Union-Find data structure
        parent = {i: i for i in range(len(self.image_paths))}

        def find(x):
            while parent[x] != x:
                parent[x] = parent[parent[x]]
                x = parent[x]
            return x

        def union(x, y):
            parent[find(x)] = find(y)

        # Find neighbors for each image
        distances, indices = self.nn_model.kneighbors(self.embeddings)

        # Connect similar images
        for i in range(len(self.image_paths)):
            for j, dist in zip(indices[i], distances[i]):
                if i != j and dist < (1 - self.threshold):  # cosine similarity threshold
                    union(i, j)

        # Group images by connected components
        groups = defaultdict(list)
        for i in range(len(self.image_paths)):
            root = find(i)
            groups[root].append(self.image_paths[i])

        # Filter groups with duplicates
        duplicate_groups = {k: v for k, v in groups.items() if len(v) > 1}

        print(f"Found {len(duplicate_groups)} duplicate groups")
        for i, (key, group) in enumerate(list(duplicate_groups.items())[:5]):
            print(f"Group {i+1}: {len(group)} images - {group[:3]}...")

        return duplicate_groups

    def detect_duplicates_in_dataset(self, df_list, df_names):
        """Detect duplicates across all datasets"""
        all_files = []
        all_dfs = []
        df_indices = []

        # Collect all files
        for df, name in zip(df_list, df_names):
            files = df['file_name'].tolist()
            all_files.extend(files)
            all_dfs.extend([name] * len(files))
            df_indices.extend(range(len(files)))

        # Build embeddings for all files
        self.build_embeddings(IMAGES_ROOT, all_files)

        # Find duplicate groups
        duplicate_groups = self.find_duplicate_groups()

        # Map groups to original dataframes
        file_to_df = dict(zip(all_files, all_dfs))
        file_to_idx = dict(zip(all_files, df_indices))

        return duplicate_groups, file_to_df, file_to_idx


# =====================================================
# APPLY DUPLICATE DETECTION
# =====================================================
print("=" * 60)
print("DUPLICATE DETECTION AND DATA SPLITTING")
print("=" * 60)

detector = DuplicateDetector(threshold=0.85, device=device)
duplicate_groups, file_to_df, file_to_idx = detector.detect_duplicates_in_dataset(
    [train_df, val_df, test_df],
    ['train', 'val', 'test']
)

# Force all duplicates into training set
print("\nMoving duplicates to training set...")
moved_count = 0

# Create copies to avoid modifying during iteration
train_df_updated = train_df.copy()
val_df_updated = val_df.copy()
test_df_updated = test_df.copy()

for group_id, files in duplicate_groups.items():
    group_dfs = [file_to_df[f] for f in files]

    # If group spans multiple splits, move all to train
    if len(set(group_dfs)) > 1:
        for file in files:
            original_split = file_to_df[file]
            if original_split != 'train':
                if original_split == 'val':
                    file_data = val_df_updated[val_df_updated['file_name'] == file]
                    if len(file_data) > 0:
                        val_df_updated = val_df_updated[val_df_updated['file_name'] != file]
                        if file not in train_df_updated['file_name'].values:
                            train_df_updated = pd.concat([train_df_updated, file_data],
                                                         ignore_index=True)
                            moved_count += 1

                elif original_split == 'test':
                    file_data = test_df_updated[test_df_updated['file_name'] == file]
                    if len(file_data) > 0:
                        test_df_updated = test_df_updated[test_df_updated['file_name'] != file]
                        if file not in train_df_updated['file_name'].values:
                            train_df_updated = pd.concat([train_df_updated, file_data],
                                                         ignore_index=True)
                            moved_count += 1

print(f"Moved {moved_count} images to training set to avoid data leakage")

# Update the dataframes
train_df = train_df_updated
val_df = val_df_updated
test_df = test_df_updated

print(f"Final dataset sizes:")
print(f"Train: {len(train_df)} images")
print(f"Val: {len(val_df)} images")
print(f"Test: {len(test_df)} images")

# =====================================================
# ENHANCED DATASET CLASS WITH PROPER SAMPLER FIX
# =====================================================
class DisasterDataset(Dataset):
    def __init__(self, df, directory, transform=None):
        self.df = df.reset_index(drop=True)
        self.dir = directory
        self.transform = transform
        
        # Filter out missing files and create mapping
        self.valid_indices = []
        self.valid_files = []
        self.valid_labels = []
        
        for idx in range(len(self.df)):
            fname = self.df.loc[idx, "file_name"]
            img_path = os.path.join(self.dir, fname)
            if os.path.exists(img_path):
                self.valid_indices.append(idx)
                self.valid_files.append(fname)
                self.valid_labels.append(self.df.loc[idx, "class_idx"])
            else:
                print(f"Warning: Image not found - {img_path}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        fname = self.valid_files[idx]
        label = self.valid_labels[idx]

        img_path = os.path.join(self.dir, fname)
        img = Image.open(img_path).convert("RGB")
        if self.transform:
            img = self.transform(img)

        return img, label

# =====================================================
# STRONGER TRANSFORMS WITH HEAVY AUGMENTATION
# =====================================================
train_tf = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.5, 1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.3),
    transforms.RandomApply([
        transforms.ColorJitter(0.4, 0.4, 0.4, 0.2)
    ], p=0.8),
    transforms.RandomRotation(20),
    transforms.RandomGrayscale(p=0.2),
    transforms.RandomApply([
        transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 2.0))
    ], p=0.3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.4, scale=(0.02, 0.3), value='random')
])

test_tf = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

train_ds = DisasterDataset(train_df, TRAIN_DIR, train_tf)
val_ds = DisasterDataset(val_df, VAL_DIR, test_tf)
test_ds = DisasterDataset(test_df, TEST_DIR, test_tf)

print(f"Valid training samples: {len(train_ds)}")
print(f"Valid validation samples: {len(val_ds)}")
print(f"Valid test samples: {len(test_ds)}")

# =====================================================
# CLASS WEIGHTED SAMPLER FOR IMBALANCED DATA (FIXED)
# =====================================================
# Use the actual labels from the filtered dataset
train_labels = train_ds.valid_labels
class_weights = compute_class_weight('balanced',
                                     classes=np.unique(train_labels),
                                     y=train_labels)
sample_weights = [class_weights[label] for label in train_labels]

# Create sampler with the correct length
sampler = WeightedRandomSampler(sample_weights,
                                len(sample_weights),  # Use actual filtered dataset size
                                replacement=True)

# On Windows, num_workers=0 avoids multiprocessing issues
loader_kwargs = {
    "batch_size": 32,
    "num_workers": 0,
    "pin_memory": (device.type == "cuda")
}

train_loader = DataLoader(train_ds, sampler=sampler, **loader_kwargs)
val_loader = DataLoader(val_ds, shuffle=False, **loader_kwargs)
test_loader = DataLoader(test_ds, shuffle=False, **loader_kwargs)

# =====================================================
# MULTIMODAL MODEL WITH REGULARIZATION
# =====================================================
class MultimodalDisasterModel(nn.Module):
    def __init__(self, num_classes, dropout_rate=0.6):
        super().__init__()

        # Image backbone
        self.image_backbone = models.densenet121(
            weights=models.DenseNet121_Weights.IMAGENET1K_V1
        )

        # Freeze early layers
        for param in list(self.image_backbone.parameters())[:100]:
            param.requires_grad = False

        # Replace classifier
        in_features = self.image_backbone.classifier.in_features
        self.image_backbone.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(in_features, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
        )

        # Final classifier
        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate * 0.7),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        image_features = self.image_backbone(x)
        output = self.classifier(image_features)
        return output

    def unfreeze_more_layers(self, epoch):
        if epoch > 8:
            for param in list(self.image_backbone.parameters())[-100:]:
                param.requires_grad = True
        if epoch > 15:
            for param in self.image_backbone.parameters():
                param.requires_grad = True


model = MultimodalDisasterModel(len(class_names), dropout_rate=0.6)
model = model.to(device)

# Label smoothing loss
class_weights_tensor = torch.tensor(class_weights,
                                    dtype=torch.float32,
                                    device=device)


class LabelSmoothCrossEntropyLoss(nn.Module):
    def __init__(self, weight=None, smoothing=0.1):
        super().__init__()
        self.weight = weight
        self.smoothing = smoothing

    def forward(self, pred, target):
        log_probs = torch.nn.functional.log_softmax(pred, dim=-1)
        nll_loss = -log_probs.gather(dim=-1, index=target.unsqueeze(1))
        nll_loss = nll_loss.squeeze(1)
        smooth_loss = -log_probs.mean(dim=-1)
        loss = (1 - self.smoothing) * nll_loss + self.smoothing * smooth_loss
        if self.weight is not None:
            loss = loss * self.weight[target]
        return loss.mean()


criterion = LabelSmoothCrossEntropyLoss(weight=class_weights_tensor,
                                        smoothing=0.1)
optimizer = optim.AdamW(model.parameters(),
                        lr=5e-5,
                        weight_decay=5e-4)
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
    optimizer,
    T_0=10,
    T_mult=1,
    eta_min=1e-6
)
max_grad_norm = 0.5

# =====================================================
# TRAINING LOOP
# =====================================================
train_losses = []
val_losses = []
train_accs = []
val_accs = []

BEST_VAL_ACC = 0.0
BEST_VAL_LOSS = float('inf')
BEST_PTH_PATH = "best_multimodal_model.pth"
BEST_PKL_PATH = "best_multimodal_model.pkl"
PATIENCE = 10
epochs_no_improve = 0


def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total, correct, loss_sum = 0, 0, 0

    pbar = tqdm(loader, desc="Training", leave=False)
    for X, y in pbar:
        X = X.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        optimizer.zero_grad()

        with autocast('cuda', enabled=use_amp):
            out = model(X)
            loss = criterion(out, y)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)
        scaler.step(optimizer)
        scaler.update()

        loss_sum += loss.item() * X.size(0)
        preds = torch.argmax(out, 1)
        correct += (preds == y).sum().item()
        total += y.size(0)

        pbar.set_postfix({'loss': loss.item(), 'acc': correct / total})

    return loss_sum / total, correct / total


def evaluate_epoch(model, loader, criterion, device):
    model.eval()
    total, correct, loss_sum = 0, 0, 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for X, y in loader:
            X = X.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)

            with autocast('cuda', enabled=use_amp):
                out = model(X)
                loss = criterion(out, y)

            loss_sum += loss.item() * X.size(0)
            preds = torch.argmax(out, 1)
            correct += (preds == y).sum().item()
            total += y.size(0)

            all_preds.extend(preds.detach().cpu().numpy())
            all_labels.extend(y.detach().cpu().numpy())

    return loss_sum / total, correct / total, all_preds, all_labels


print("Starting multimodal training with duplicate-aware splits...")
for epoch in range(40):
    model.unfreeze_more_layers(epoch)

    train_loss, train_acc = train_epoch(model, train_loader, optimizer, criterion, device)
    val_loss, val_acc, _, _ = evaluate_epoch(model, val_loader, criterion, device)

    scheduler.step(epoch)

    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)

    current_lr = optimizer.param_groups[0]['lr']
    print(
        f"Epoch {epoch + 1:2d} | "
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | "
        f"LR: {current_lr:.2e}"
    )

    improvement = val_acc > BEST_VAL_ACC + 0.001 or val_loss < BEST_VAL_LOSS - 0.002
    if improvement:
        BEST_VAL_ACC = max(BEST_VAL_ACC, val_acc)
        BEST_VAL_LOSS = min(BEST_VAL_LOSS, val_loss)
        epochs_no_improve = 0

        torch.save(model.state_dict(), BEST_PTH_PATH)
        with open(BEST_PKL_PATH, 'wb') as f:
            pickle.dump(model, f)
        print("  ↳ ✅ Saved new best model!")
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= PATIENCE:
            print(f"🛑 Early stopping after {epoch + 1} epochs!")
            break

    if epoch > 10 and (train_acc - val_acc) > 0.15:
        print(f"🛑 Stopping due to severe overfitting (gap: {train_acc - val_acc:.3f})")
        break

# =====================================================
# SAVE RESULTS
# =====================================================
training_results = {
    'model_state_dict': model.state_dict(),
    'class_names': class_names,
    'class2idx': class2idx,
    'idx2class': idx2class,
    'train_losses': train_losses,
    'val_losses': val_losses,
    'train_accs': train_accs,
    'val_accs': val_accs,
    'best_val_acc': BEST_VAL_ACC,
    'best_val_loss': BEST_VAL_LOSS,
    'duplicate_groups': duplicate_groups
}

with open("complete_training_results.pkl", 'wb') as f:
    pickle.dump(training_results, f)

# =====================================================
# PLOTS
# =====================================================
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label="Train Loss", linewidth=2, alpha=0.8)
plt.plot(val_losses, label="Val Loss", linewidth=2, alpha=0.8)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve - Duplicate Aware')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(train_accs, label="Train Acc", linewidth=2, alpha=0.8)
plt.plot(val_accs, label="Val Acc", linewidth=2, alpha=0.8)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve - Duplicate Aware')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("training_curves_duplicate_aware.png", dpi=150, bbox_inches='tight')
plt.show()

# =====================================================
# TEST EVALUATION
# =====================================================
print("\n" + "=" * 60)
print("FINAL TEST EVALUATION")
print("=" * 60)

model.load_state_dict(torch.load(BEST_PTH_PATH, map_location=device))
model.to(device)
model.eval()

test_loss, test_acc, y_true, y_pred = evaluate_epoch(model, test_loader, criterion, device)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Save test results
test_results = {
    'y_true': y_true,
    'y_pred': y_pred,
    'test_accuracy': test_acc,
    'test_loss': test_loss
}

with open("test_results.pkl", 'wb') as f:
    pickle.dump(test_results, f)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix - Duplicate Aware')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("confusion_matrix_duplicate_aware.png", dpi=150, bbox_inches='tight')
plt.show()

# =====================================================
# FINAL SUMMARY
# =====================================================
final_gap = train_accs[-1] - val_accs[-1] if train_accs and val_accs else 0.0
print(f"\n" + "=" * 60)
print("TRAINING SUMMARY")
print("=" * 60)
print(f"✅ Duplicate groups found: {len(duplicate_groups)}")
print(f"✅ Images moved to train: {moved_count}")
print(f"✅ Best validation accuracy: {BEST_VAL_ACC:.4f}")
print(f"✅ Test accuracy: {test_acc:.4f}")
print(f"✅ Final overfitting gap: {final_gap:.4f}")

print(f"\n📁 FILES SAVED:")
print(f"  - {BEST_PTH_PATH} (PyTorch model)")
print(f"  - {BEST_PKL_PATH} (Pickle model)")
print(f"  - complete_training_results.pkl (Training data)")
print(f"  - test_results.pkl (Test results)")
print(f"  - training_curves_duplicate_aware.png (Plots)")
print(f"  - confusion_matrix_duplicate_aware.png (Confusion matrix)")

print(f"\n🎯 DUPLICATE HANDLING:")
print(f"  - Threshold: 0.85")
print(f"  - Groups spanning splits: "
      f"{len([g for g in duplicate_groups.values() if len(set(file_to_df[f] for f in g)) > 1])}")
print(f"  - Data leakage prevented: ✅")
```
